# 8. Control Statements (Break, Continue, Pass) 

#### ‚Ä¢ Practical Example: 1) Write a Python program to skip 'banana' in a list using the continue 
####   statement. List1 = ['apple', 'banana', 'mango'] 

```python
# list of Fruits
list1 = ['apple', 'banana', 'mango']

for fruit in list1:
    if fruit=='banana':
        continue   # banana skip
    print(fruit)
```

#### Practical Example: 2) Write a Python program to stop the loop once 'banana' is found using 
#### the break statement. 

```python
list1 = ['apple', 'banana', 'mango']

for fruit in list1:
    if fruit == 'banana':
        break
    print(fruit)
```

# 9. String Manipulation 

#### Que.1 Write a Python program to demonstrate string slicing.

```python
# user input string 

string = input("Enter the String: ")

# Accesing first 5 character 
print("String first 5 character : ",string[:5])

# Accesing character index 2 to 6:
print("chracter index 2 to 6: ",string[2:6])

# 3. Accessing string from index 3 till the end
print("string from index 3 to end :",string[3:])

# 4. Accessing string from beginning to second last character
print("Up to second last character: ",string[:-1])

# 5. Accesing every second character 
print("String every second character : ",string[::2])

# 6. Reversing the string 
print(string,"String Reversed:",string[::-1])
```

#### Que.2 Write a Python program that manipulates and prints strings using various string methods. 

```python
# Taking input from the user
text = input("Enter a string: ")

print("\nOriginal String:", text)

# 1. Convert to uppercase
print("Uppercase:", text.upper())

# 2. Convert to lowercase
print("Lowercase:", text.lower())

# 3. Capitalize first character
print("Capitalized:", text.capitalize())

# 4. Title case (first letter of each word capitalized)
print("Title Case:", text.title())

# 5. Count occurrences of a substring
substring = input("\nEnter a substring to count: ")
print(f"Count of '{substring}':", text.count(substring))

# 6. Replace a substring with another
old = input("\nEnter substring to replace: ")
new = input("Enter new substring: ")
print("After replacement:", text.replace(old, new))

# 7. Strip leading and trailing spaces
print("\nString with spaces stripped:", text.strip())

# 8. Split the string into words
print("Split into words:", text.split())

# 9. Join words with a hyphen
words = text.split()
print("Joined with hyphen:", "-".join(words))

# 10. Reverse the string
print("Reversed string:", text[::-1])
```
